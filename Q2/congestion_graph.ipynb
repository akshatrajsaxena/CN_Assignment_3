```python
import matplotlib.pyplot as plt

def plot_congestion_window(filename):
    times = []
    window_sizes = []
    
    with open(filename, 'r') as file:
        for line in file:
            if line.strip() and not line.startswith('#'):
                try:
                    time, window_size = map(float, line.strip().split())
                    times.append(time)
                    window_sizes.append(window_size)
                except ValueError:
                    continue
    
    plt.figure(figsize=(12, 6))
    plt.plot(times, window_sizes, marker='o', linestyle='-', markersize=4)
    
    plt.title('TCP Congestion Window Size Over Time', fontsize=16)
    plt.xlabel('Time (seconds)', fontsize=12)
    plt.ylabel('Congestion Window Size', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)
    
    plt.tight_layout()
    plt.savefig('congestion_window_plot.png', dpi=300)
    plt.show()
plot_congestion_window('tcp-example.cwnd')
```
